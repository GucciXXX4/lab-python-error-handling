````python
# initialize inventory with error handling
def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_input = False
        while not valid_input:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity >= 0:
                    inventory[product] = quantity
                    valid_input = True
                else:
                    print("Quantity cannot be negative. Please enter a valid quantity.")
            except ValueError:
                print("Invalid input. Please enter a number.")
    return inventory


# get customer orders with error handling
def get_customer_orders(products, inventory):
    customer_orders = set()

    # number of orders 
    while True:
        try:
            num_orders = int(input("Enter the number of customer orders: "))
            if num_orders >= 0:
                break
            else:
                print("Number of orders cannot be negative.")
        except ValueError:
            print("Invalid input. Please enter a number.")

    # take the product names 
    for i in range(num_orders):
        while True:
            product_choice = input(f"Enter product {i+1}: ").lower()
            if product_choice in products:
                if inventory.get(product_choice, 0) > 0:
                    customer_orders.add(product_choice)
                    break
                else:
                    print(f"Sorry, {product_choice} is out of stock. Please choose another product.")
            else:
                print("Invalid product name. Please choose from:", products)

    return customer_orders


# update inventory removes 0 items
def update_inventory(customer_orders, inventory):
    for product in customer_orders:
        inventory[product] -= 1
    inventory = {p: q for p, q in inventory.items() if q > 0}
    return inventory


# calculate order statistics
def calculate_order_statistics(customer_orders, products):
    total_products_ordered = len(customer_orders)
    total_products_available = len(products)
    percentage_ordered = (total_products_ordered / total_products_available) * 100
    return (total_products_ordered, percentage_ordered)


# print order statistics
def print_order_statistics(order_statistics):
    print("\nOrder Statistics:")
    print(f"Total Products Ordered: {order_statistics[0]}")
    print(f"Percentage of Unique Products Ordered: {order_statistics[1]:.2f}%")


# show new inventory
def print_updated_inventory(inventory):
    print("\nUpdated Inventory:")
    for product in inventory:
        print(f"{product}: {inventory[product]}")


# calculate total price with error handling
def calculate_total_price(customer_orders):
    prices = {}
    for product in customer_orders:
        while True:
            try:
                price = float(input(f"Enter the price of {product}: "))
                if price >= 0:
                    prices[product] = price
                    break
                else:
                    print("Price cannot be negative. Try again.")
            except ValueError:
                print("Invalid input. Please enter a number.")
    total_price = sum(prices.values())
    return total_price


# program
def main():
    products = ["t-shirt", "mug", "hat", "book", "keychain"]

    # initialize inventory
    inventory = initialize_inventory(products)

    # get customer orders
    customer_orders = get_customer_orders(products, inventory)

    # calculate statistics
    order_statistics = calculate_order_statistics(customer_orders, products)
    print_order_statistics(order_statistics)

    # new inventory
    inventory = update_inventory(customer_orders, inventory)
    print_updated_inventory(inventory)

    # calculate and show total price
    total_price = calculate_total_price(customer_orders)
    print(f"\nTotal Price: {total_price:.2f}")


if __name__ == "__main__":
    main()
````
